```python
import numpy as np
import pandas as pd 
from random import sample
```

```python
baseFloor_df=pd.read_csv('Stadium_mod/costmap5.csv', sep=',',header=None)
```

```python
baseFloor = baseFloor_df.to_numpy()

valid_pairs = []

for i in range(len(baseFloor)):
    for j in range(len(baseFloor[0])):
        if baseFloor[i,j] != -1:
            valid_pairs.append([i,j])
len(valid_pairs) * .4
```

```python
allCostChanges = []


start_time = 5
end_time = 105
time_step = 5
floor_number = 0
costDelta = 200

currentFloor = baseFloor.copy()

for t in range(start_time,end_time,time_step):
    lastFloor = currentFloor.copy()
    currentFloor = baseFloor.copy()
    random_pairs = sample(valid_pairs,1000)
    for pair in random_pairs:
        i = pair[0]
        j = pair[1]
        currentFloor[i][j] = costDelta
        if lastFloor[i][j] != currentFloor[i][j]:
            val = currentFloor[i][j]
            costChange = [t,floor_number,i,j,val]
            allCostChanges.append(costChange)
```

```python
pd.DataFrame(allCostChanges).to_csv("Stadium/costChanges_floor0.csv")
```

```python
def makeCostChangeCSV(floor_number):

    
    baseFloor_df=pd.read_csv('Stadium_mod/costmap' + str(floor_number) + '.csv', sep=',',header=None)
    
    baseFloor = baseFloor_df.to_numpy()
    
    valid_pairs = []

    for i in range(len(baseFloor)):
        for j in range(len(baseFloor[0])):
            if baseFloor[i,j] != -1 and baseFloor[i,j] != 1000:
                valid_pairs.append([i,j])
                
    allCostChanges = []

    start_time = 5
    end_time = 105
    time_step = 5
    
    z = floor_number*2
    costDelta = 200
    
    pct_changed = 0.3

    currentFloor = baseFloor.copy()

    for t in range(start_time,end_time,time_step):
        lastFloor = currentFloor.copy()
        currentFloor = baseFloor.copy()
        random_pairs = sample(valid_pairs, int(pct_changed * len(valid_pairs)))
        
        for pair in random_pairs:
                
            i = pair[0]
            j = pair[1]
            currentFloor[i][j] = costDelta
            if lastFloor[i][j] != currentFloor[i][j]:
                val = currentFloor[i][j]
                costChange = [t,z,i,j,val]
                allCostChanges.append(costChange)
        print("t=",t)
        pd.DataFrame(currentFloor).to_csv("Stadium_mod/costMapChanged" + str(floor_number*2) + ".csv")
    return allCostChanges
    #pd.DataFrame(allCostChanges).to_csv("Stadium/costChanges_floor" + str(floor_number) + ".csv")
    
    
```

```python
allCostChanges = makeCostChangeCSV(0)
for i in range(1,6):
    currentCostChanges = makeCostChangeCSV(i)
    allCostChanges = np.vstack((allCostChanges, currentCostChanges))

```
